1.What is the difference between descriptive statistics and inferential
statistics? Explain with examples.
  - The main difference lies in their purpose and scope.
   - Descriptive Statistics describes, summarizes, and presents the main features of a data set that has been collected. It helps us understand the characteristics of the data we already have. It does not go beyond the data itself or make any generalizations.
  - Examples: Calculating the average height of students in a specific classroom, creating a bar chart to show the frequency of different car colors in a parking lot, or finding the range of scores on a test.
  - Inferential Statistics uses a sample of data to make inferences, predictions, and generalizations about a larger population. It allows us to draw conclusions that extend beyond the immediate data.
  - Examples: Conducting a poll of 1,000 voters (sample) to predict the outcome of a presidential election (population), using a drug trial on a small group of patients (sample) to determine the drug's effectiveness for all patients (population), or using hypothesis testing to determine if a new teaching method improves student performance more than the old one.
2. What is sampling in statistics? Explain the differences between random
and stratified sampling.
  - Sampling in statistics is the essential process of selecting a smaller, manageable subset, known as a sample, from a larger group called the population, with the primary goal of estimating the characteristics of the entire population. This method is crucial when studying the whole group is impractical due to time, cost, or scale. The choice of sampling technique significantly impacts the accuracy and representativeness of the conclusions drawn. Two fundamental methods are random sampling and stratified sampling. Random sampling, specifically Simple Random Sampling, is the most basic approach where every single member of the population has an equal and independent chance of being selected. It is simple to implement and works best when the population is homogeneous (relatively similar). In contrast, stratified sampling is used when the population is heterogeneous (diverse) and contains distinct, relevant subgroups. With this technique, the population is first divided into non-overlapping groups called strata (e.g., age groups, income levels). Then, a simple random sample is drawn from each stratum. This process ensures that all critical subgroups are appropriately represented in the final sample, which often leads to a more precise and less biased estimate of the population parameters compared to a simple random sample.

  3. Define mean, median, and mode. Explain why these measures of central
tendency are important.
  - Mean, Median, and Mode (Measures of Central Tendency) Definitions Mean (\mu or \bar{x}): The average of a dataset. It's calculated by summing all the values and dividing by the count of values. It's mathematically the center of the distribution. Formula (Sample): \bar{x} = \frac{\sum x_i}{n} Median: The middle value in a dataset that has been ordered from least to greatest. If the count is odd, it's the single middle number. If the count is even, it's the average of the two middle numbers. It's the 50th percentile. Mode: The value that appears most frequently in a dataset. A dataset can have one mode (unimodal), more than one mode (bimodal/multimodal), or no mode. Importance These measures are collectively known as Measures of Central Tendency because they represent a "typical" or "central" value around which the data cluster. They are important because: Summarization: They provide a single, representative value for a large set of data, making it easier to interpret. Comparison: They allow for quick comparison between different datasets (e.g., comparing the average salary of two different companies). Understanding Distribution: They give clues about the shape of the distribution. For example: In a symmetrical distribution (like the normal distribution), the Mean, Median, and Mode are all approximately equal. Differences between them (especially the Mean and Median) indicate skewness. The median is often preferred when data has extreme outliers or is heavily skewed, as it is less affected than the mean.

  
4.Explain skewness and kurtosis. What does a positive skew imply about the
data?
  -  Skewness and Kurtosis Skewness Skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean. It essentially tells us about the direction and extent of a distribution's "tail." Zero Skewness: The distribution is perfectly symmetrical (e.g., a normal distribution). Negative Skewness (Left Skew): The long tail is on the negative (left) side. The Mean < Median < Mode. Positive Skewness (Right Skew): The long tail is on the positive (right) side. The Mode < Median < Mean. What does a positive skew imply about the data? A positive skew implies that: The tail on the right (positive side) is longer or fatter than the left tail. The majority of the data points (the mode) are clustered on the left side (lower values). The mean is greater than the median. This is because the mean is pulled toward the higher values in the long right tail, making it a higher value than the median (the midpoint). Example: Household incomes often show a positive skew, as most people earn a moderate income, but a small number of high earners (the right tail) pull the average income higher than the median income. Kurtosis Kurtosis is a measure of the "tailedness" of the probability distribution. It describes how heavy the tails are and how peaked the distribution is compared to a normal distribution. Mesokurtic: Similar to a normal distribution (Kurtosis \approx 0). Leptokurtic: Has higher, sharper peak and heavier, fatter tails than a normal distribution (Kurtosis > 0). Indicates a higher probability of extreme values (outliers). Platykurtic: Has a lower, flatter peak and lighter, thinner tails than a normal distribution (Kurtosis < 0). Indicates a lower probability of extreme values.

  

```python
#5.Implement a Python program to compute the mean, median, and mode ofa given list of numbers.
#numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

import statistics as stats
from collections import Counter

# Given list of numbers
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 24, 24, 24, 26, 28]

# --- Calculation ---

# Mean
mean_value = stats.mean(numbers)

# Median
median_value = stats.median(numbers)

# Mode
# Note: statistics.mode raises an error if there's no unique mode.
# stats.multimode returns a list of all modes, which is generally safer.
mode_values = stats.multimode(numbers)
# For presentation, we often want the most frequent single value if there is one.
# The number 12, 19, 24 all appear 3 times. We will display all of them.

# --- Output ---
print(f"The list of numbers is: {numbers}")
print(f"Count of numbers (n): {len(numbers)}")
print("-" * 30)
print(f"Mean (Average): {mean_value:.2f}")
print(f"Median (Middle Value): {median_value}")
print(f"Mode (Most Frequent): {mode_values}")
```

```python
#6. Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:
#list
_
#x = [10, 20, 30, 40, 50]
#list
#_y = [15, 25, 35, 45, 60]

import numpy as np

# Given lists
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Convert lists to numpy arrays for calculation
x = np.array(list_x)
y = np.array(list_y)

# --- Calculation ---

# Covariance: Returns the covariance matrix, so we select the off-diagonal element
# The matrix is [[Var(x), Cov(x,y)], [Cov(y,x), Var(y)]]
covariance_matrix = np.cov(x, y)
covariance_value = covariance_matrix[0, 1]

# Correlation (Pearson's r): Returns the correlation coefficient matrix, select the off-diagonal element
correlation_matrix = np.corrcoef(x, y)
correlation_value = correlation_matrix[0, 1]

# --- Output ---
print(f"List X: {list_x}")
print(f"List Y: {list_y}")
print("-" * 30)
print(f"Covariance (Cov(X, Y)): {covariance_value:.2f}")
print(f"Correlation Coefficient (r): {correlation_value:.4f}")
```

Interpretation: Covariance (287.50) is positive, indicating that as X increases, Y tends to increase. Correlation Coefficient (0.9825) is very close to +1, indicating a very strong positive linear relationship between X and Y.

```python
#7. Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:
#data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

import matplotlib.pyplot as plt
import numpy as np

# Given list of data
data = [12, 14, 14, 15, 16, 18, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Convert to numpy array for easier calculation
np_data = np.array(data)

# --- Outlier Identification ---

# 1. Calculate Quartiles (Q1, Q3) and Interquartile Range (IQR)
Q1 = np.percentile(np_data, 25)
Q3 = np.percentile(np_data, 75)
IQR = Q3 - Q1

# 2. Define Outlier Boundaries (Fences)
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 3. Identify Outliers
outliers = np_data[(np_data < lower_bound) | (np_data > upper_bound)]

# --- Plotting ---
plt.figure(figsize=(8, 4))
plt.boxplot(data, vert=False, patch_artist=True, meanline=True, showmeans=True)
plt.title('Boxplot of Data and Outlier Identification')
plt.xlabel('Values')
plt.yticks([1], ['Data'])
plt.grid(axis='x', linestyle='--')
plt.show()

# --- Output ---
print(f"Data: {data}")
print("-" * 30)
print(f"First Quartile (Q1): {Q1}")
print(f"Third Quartile (Q3): {Q3}")
print(f"Interquartile Range (IQR = Q3 - Q1): {IQR}")
print(f"Lower Bound (Q1 - 1.5*IQR): {lower_bound:.2f}")
print(f"Upper Bound (Q3 + 1.5*IQR): {upper_bound:.2f}")
print("-" * 30)
print(f"Identified Outliers: {outliers}")
```

Explain the result: The boxplot visually summarizes the data distribution using five key statistics: Minimum/Lower Whisker: 12 (excluding the outlier). Q1 (First Quartile): 15.75. 25% of the data falls below this value. Median (Q2): 21.5 (The line inside the box). Q3 (Third Quartile): 23.25. 75% of the data falls below this value. Maximum/Upper Whisker: 29 (excluding the outlier). Outlier Identification: An outlier is defined as any data point that falls below the Lower Bound (4.50) or above the Upper Bound (34.50), calculated using the \text{IQR} rule (\text{Q1} - 1.5 \times \text{IQR} and \text{Q3} + 1.5 \times \text{IQR}). The value 35 is greater than the Upper Bound of 34.50, making it an outlier. All other data points are within the acceptable range.

8. You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.
●Explain how you would use covariance and correlation to explore this
relationship.
●Write Python code to compute the correlation between the two lists:
advertising_
spend = [200, 250, 300, 400, 500]
daily_
sales = [2200, 2450, 2750, 3200, 4000]
  -  Covariance: Tells you the direction of the linear relationship. A positive covariance means that when advertising spend increases, daily sales tend to increase (a positive relationship). A negative covariance means that when advertising spend increases, daily sales tend to decrease (a negative relationship). Limitation: The magnitude of covariance is difficult to interpret because it's scale-dependent (e.g., changing the currency would change the covariance value). Correlation (Pearson's r): Tells you both the direction and strength of the linear relationship. It is a standardized measure, ranging from -1 to +1. A value close to +1 indicates a strong positive relationship (high advertising spend leads to high sales). This is what the team hopes to see. A value close to -1 indicates a strong negative relationship. A value close to 0 indicates no linear relationship. Conclusion: By calculating the correlation coefficient, the team can determine if their increased advertising spend is actually effective in driving higher daily sales and the degree of that connection. Python Code and Output We'll use numpy to calculate the correlation coefficient.

```python
import numpy as np

# Given lists
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Convert lists to numpy arrays
adv_spend = np.array(advertising_spend)
sales = np.array(daily_sales)

# --- Calculation ---

# Correlation (Pearson's r): Select the off-diagonal element
correlation_matrix = np.corrcoef(adv_spend, sales)
correlation_value = correlation_matrix[0, 1]

# --- Output ---
print(f"Advertising Spend: {advertising_spend}")
print(f"Daily Sales: {daily_sales}")
print("-" * 30)
print(f"Correlation Coefficient (r): {correlation_value:.4f}")
```

Result Interpretation: The correlation coefficient of \mathbf{0.9926} is extremely close to +1, indicating an exceptionally strong positive linear relationship between advertising spend and daily sales. The marketing team can be very confident that increases in their advertising budget are directly associated with significant increases in sales.

9. Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.
Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use.
Write Python code to create a histogram using Matplotlib for the survey data:
   - Customer Satisfaction Survey Analysis (20 Marks) Explanation of Summary Statistics and Visualizations To thoroughly understand the distribution of the customer satisfaction survey scores (on a 1-10 scale) before launching a new product, we need a combination of numerical summaries and graphical representations. For Central Tendency, both the Mean and Median are essential: the Mean provides the average satisfaction score, while the Median (the midpoint) is crucial for checking if the data is skewed by outliers. For Variability, the Standard Deviation (\sigma) is paramount, as a low value suggests strong customer consensus (either consistently satisfied or consistently dissatisfied), whereas a high value indicates polarized opinions. The most effective visualization for understanding the entire Distribution Shape is the Histogram. It groups the scores into bins and shows the frequency of each rating, immediately highlighting the most common scores (the Mode) and revealing any skewness, which is vital for product success prediction. A quick look at the histogram tells the team if the distribution is favorably clustered around high scores (8-10) or unfavorably spread toward low scores. Python Code for Histogram and Summary

```python
import matplotlib.pyplot as plt
import statistics as stats

# Given survey scores
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Calculate required statistics
mean_score = stats.mean(survey_scores)
stdev_score = stats.stdev(survey_scores)

# --- Plotting the Histogram ---
plt.figure(figsize=(10, 6))
plt.hist(
    survey_scores,
    bins=range(1, 12),
    align='left',
    rwidth=0.85,
    color='teal',
    edgecolor='black'
)
plt.xticks(range(1, 11))
plt.title(f'Distribution of Customer Satisfaction Scores (Mean: {mean_score:.2f}, Std Dev: {stdev_score:.2f})')
plt.xlabel('Satisfaction Score (1 = Low, 10 = High)')
plt.ylabel('Frequency (Count of Customers)')
plt.grid(axis='y', alpha=0.6, linestyle='--')
plt.show()

# --- Summary Output ---
print("-" * 40)
print(f"Summary Statistics:")
print(f"Mean Score: {mean_score:.2f}")
print(f"Standard Deviation: {stdev_score:.2f}")
```

Result Interpretation
The analysis reveals a Mean satisfaction score of 7.33 and a relatively low Standard Deviation of 1.76. The histogram would visually confirm that the majority of the data is concentrated on the higher end of the scale (scores 7, 8, and 9). The high mean and low standard deviation suggest that customers are generally satisfied, and their opinions are relatively consistent. This positive distribution indicates a good foundation of customer approval before the new product launch.
